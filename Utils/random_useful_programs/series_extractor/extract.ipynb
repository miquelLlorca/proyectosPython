```python
in_path = '/home/miquel/proyectosPython/random/series_extractor/titles.txt'
with open(in_path,'r') as series_file:
    series = series_file.read().lower().split('\n')
series
```

```python
series_set = set()

for title in series:
    start_index = title.find("capitulo")
    if(start_index != -1):
        series_set.add(title[:start_index])
    else:
        start_index = title.find("1x")
        if(start_index != -1):
            series_set.add(title[:start_index])
     
     
series_set
```
